## Introduction


The COVID-19 pandemic drastically shifted work environments, with remote work becoming the new norm. While it offered flexibility, it also raised concerns about its long-term impact on employee health. <br>This study analyzes data from 2025 to explore how remote work has affected physical activity, mental health, sleep quality, screen time, and job satisfaction. By examining these factors, the report aims to identify patterns that reveal both positive and negative health outcomes of working remotely. <br>The insights will help employers and policymakers create healthier, more balanced remote work strategies in the post-pandemic era.


##  Problem Statement:
In recent years, mental health challenges and burnout have become increasingly prevalent across industries, especially in high-demand job roles such as data analysis and business analysis. Organizations are striving to understand the key factors contributing to employee stress, burnout, and work-life imbalance to design healthier work environments.

You have been provided with survey data from employees across various regions and industries. The dataset includes variables such as age, gender, job role, work arrangement, hours worked per week, mental health status, burnout level, physical health issues, salary range, and social isolation scores.

### Make the age range

```python
bins=[21,30,40,50,60]
labels=['22-30','31-40','41-50','51-60']

df['Age_Group']=pd.cut(df['Age'],bins=bins,labels=labels,right=True)
df['Age_Group']
```

```python
df['Age_Group'] = df['Age_Group'].cat.add_categories('Unknown')
df['Age_Group'].fillna('Unknown', inplace=True)
```

## Job role vs Burnout

This pivot table allows easy comparison of burnout levels across different job roles, which is helpful for further percentage calculations or visualizations.

```python
burnout_df = pd.pivot_table(df, index='Job_Role', columns='Burnout_Level', aggfunc='size', fill_value=0)
burnout_df.head()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(burnout_df,fmt='d',cmap='YlOrRd',linewidths=0.5,linecolor='gray',annot=True)
plt.title("burnout level heatmap by job role")
plt.ylabel("job Role")
plt.xlabel("Burnout level")
plt.tight_layout()
plt.show()
```

### 2nd method

```python
burnout_df = df.groupby('Job_Role')['Burnout_Level'].value_counts().unstack(fill_value=0)
burnout_df
```

### Convert into Percentage

```python
burnout_percent = burnout_df.div(burnout_df.sum(axis=1), axis=0) * 100
```

```python
burnout_percent.round(2)
```

```python
# Row-wise percentage
burnout_vs_mental_pct = pd.crosstab(df['Mental_Health_Status'], df['Burnout_Level'], normalize='index') * 100
print(burnout_vs_mental_pct.round(2))
```

```python
burnout_vs_mental_pct.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(8,5))
plt.title("Burnout Level Distribution by Mental Health Status")
plt.ylabel("Percentage (%)")
plt.xlabel("Mental Health Status")
plt.legend(title="Burnout Level")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
